Scrape the las table from the wikipedia page: https://en.wikipedia.org/wiki/Red_states_and_blue_states

Technologies: Beautiful Soup, Requests, Numpy, Pandas

```python
import requests
import bs4
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/Red_states_and_blue_states"
response = requests.get(url)

if (response.status_code == 200):
  soup = bs4.BeautifulSoup(response.text, "lxml");

  tables = soup.find_all("table", class_="wikitable")
  table_rows = tables[-1].find_all("tr")

  # Get all the years
  years = [item.text.replace("\n", "") for item in table_rows[0].find_all("th")]
  # print(years)

  # Get all the democratic candidats in each year
  # democrats = [item.text.replace("\n", "").strip() for item in table_rows[2].find_all("td")]
  # print(democrats)

  # Get all the republican candidates in each year
  # republicans = [item.text.replace("\n", "").strip() for item in table_rows[3].find_all("td")]
  # print(republicans)

  contents = [{years[idx] : item.text.replace("\n", "").strip() for (idx, item) in enumerate(table_rows[row_idx].find_all("td"))} for row_idx in range(2, len(table_rows))]

  # for row_idx in range(2, len(table_rows)):
  #   row_content = [item.text.replace("\n", "").strip() for item in table_rows[row_idx].find_all("td")]
  #   row_content_dict =
  #   contents.append(row_content_dict)

  print(contents[-1])
else:
  print(f"Error: {response.status_code}")
```

```python
df = pd.DataFrame(data=contents, columns=years)
df.head(10)
```
